## Решение

* Модель
  * Предлагается упростить лунный модуль до материальной точки
  * Три двигателя — вверх (сопло вниз), влево, вправо
  * Поверхность лугы предлагается сделать ступенчатой
* Графика
  * Предлагается использвоать PyGame (ведь у нас уже есть опыт!)
  * Предлагается не встраивать модель, а научиться брать из неё данные и передавать ей управление
* Переносимость и DevOps
  * Должно работать под Windows и Linux
  * Для мобильных... Вот тут надо подумать, м.б. и не PyGame
  * Ну и д.б. проект c кодом в Гите и всё такое

## Как можно научить аппарат садиться самостоятельно?

Отрастить автопилот. Именно для этого и следует отвязать модель от игры

* Аналитическое решение [задачи оптимизации](https://ru.wikipedia.org/wiki/%D0%9E%D0%BF%D1%82%D0%B8%D0%BC%D0%B8%D0%B7%D0%B0%D1%86%D0%B8%D1%8F_(%D0%BC%D0%B0%D1%82%D0%B5%D0%BC%D0%B0%D1%82%D0%B8%D0%BA%D0%B0))
  * В данном случае — [вариационного исчисления](https://ru.wikipedia.org/wiki/%D0%92%D0%B0%D1%80%D0%B8%D0%B0%D1%86%D0%B8%D0%BE%D0%BD%D0%BD%D0%BE%D0%B5_%D0%B8%D1%81%D1%87%D0%B8%D1%81%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5)
* Численное решение при помощи машинного обучения
  * [Градиентный спуск](https://ru.wikipedia.org/wiki/%D0%93%D1%80%D0%B0%D0%B4%D0%B8%D0%B5%D0%BD%D1%82%D0%BD%D1%8B%D0%B9_%D1%81%D0%BF%D1%83%D1%81%D0%BA), применительно к задачам, аналогичным поставленной — [нейронные сети](https://ru.wikipedia.org/wiki/%D0%9D%D0%B5%D0%B9%D1%80%D0%BE%D0%BD%D0%BD%D0%B0%D1%8F_%D1%81%D0%B5%D1%82%D1%8C)
  * Когда модель совсем непонятная — генетический алгоритм. Вообще это не наш случай, поэтому... давайте попробуем! =)
  * Обучающие алгоритмы будут быстро «гонять» модель, и задавать ей оптимальное управление

## Генетические алгоритмы

Введение

* Запустим на фоне Box Car 2D https://rednuht.org/genetic_cars_2/
* [Что это вообще такое?](https://dluciv.github.io/algs_and_data_structs-spbu-CB.5001/slides.html?md=11.evolution)
* Пример всяких мелких генетических (на самом деле не только) проектов https://awesomeopensource.com/projects/genetic-algorithm

Варианты исполнения

* Проще — сделать самим
* Правильнее  — разобраться с
  * https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.differential_evolution.html, или 
  * https://deap.readthedocs.io/, или
  * любым другим фреймворком

Варианты генома автопилота

* Дискретное время, битовая шкала — например, раз в 1/2 секунды включать/выключать разные двигатели
  * Лёгкие и понятные генетические операции
* Набор пауз и интервалов
  * Геном компактнее, но эволюция сложнее
